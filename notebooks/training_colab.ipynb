```python
from src.train import main as train_main

class Parameters:
    def __init__(
        self,
        base_dir: str,
        image_size: tuple,
        batch_size: int,
        epochs: int,
        seed: int,
    ):
        self.base_dir = base_dir
        self.image_size = image_size
        self.batch_size = batch_size
        self.epochs = epochs
        self.seed = seed


IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 10
SEED = 42

args = Parameters(
    base_dir=BASE_PATH,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    seed=SEED,
)

train_main(args=args)
```
